```python
def maxRow(arr):
    max_num = 0
    for x in arr:
        if( type(x) == str ):
            continue
        if( x >= max_num ):
            max_num = x
            
    return[max_num, arr.index(max_num)]


def minRow(arr):
    min_num = maxRow(arr)[0]
    for x in arr:
        if( type(x) == str or x == 0):
            continue
        if( x < min_num ):
            min_num = x
            
    return[min_num, arr.index(min_num)]


def printMatrix(arr):
    n = len(arr)
    for i in range(0,n):
        print(arr[i])


def isInArray(arr, x):
    resp = False
    for z in arr:
        if( type(z) == str ):
            continue
        if( z == x ):
            resp = True
        
    return(resp)


def maxRowCompare(arr1, arr2):
    maxc = maxRow(arr1)
    
    while isInArray(arr2, maxc[1]) == True:
        aux = arr1.copy()
        aux.remove(maxc[0])
        gde = maxRow(aux)[0]
        maxc = [gde, arr1.index(gde)]
        
    return[maxc[0], arr1.index(maxc[0])]


def maxRowCompare2(arr1, arr2):
    maxc = maxRow(arr1)
    aux = arr1.copy()
    
    while isInArray(arr2, maxc[1]) == True:
        idc = aux.index(maxc[0])
        aux[idc] = 'x'
        gde = maxRow(aux)[0]
        maxc = [gde, aux.index(gde)]

    return maxc


def routeCleaner(bucle):
    i = len(bucle)
    result = []
    
    for num in range(0,i):
        aux = bucle.copy()
        aux[num] = 'x'
        
        if isInArray(aux, bucle[num]) :
            stop = aux.index(bucle[num])
            for j in range(num, stop +1):
                bucle[j] = 'x' 
            
    for num in bucle:
        if type(num) != str:
            result.append(num)
    
    return result
        
    
def printResults(ruta, costo_min, data):
    print('Iteración -------------------------------------------------------------------')
    print('La ruta es: ',ruta ,' el flujo máximo de esta ruta es:', costo_min)

    for y in range(0, len(ruta)-1): #proceso para reajustar valores de matriz
        data[ruta[y]][ruta[y+1]] = data[ruta[y]][ruta[y+1]] - costo_min
        data[ruta[y+1]][ruta[y]] = data[ruta[y+1]][ruta[y]] + costo_min

    printMatrix(data)
    
    
def flujoMaximo(ruta, costos, fin, paso1, i):
    while isInArray(ruta, fin-1) == False:
        if not isInArray(ruta, paso1[1]) :
            #print('Inicial: ', i, ruta, costos, paso1)
            i.append(paso1[1])
            ruta.append(i[-1])
            costos.append(paso1[0])
            paso1 = maxRowCompare2(data[i[-1]], ruta)
            if paso1[1] == fin -1:
                i.append(paso1[1])
                ruta.append(i[-1])
                costos.append(paso1[0])
            #print('Final: ', i, ruta, costos, paso1)
            if paso1[0] == 0:
                jump = i.pop()
                costos.pop()
                ruta.append(paso1[1])
                ruta.append(jump)
                #print('Excepcion: ', i, ruta, costos, jump)
                paso1 = maxRowCompare2(data[i[-1]], ruta)
                if paso1[1] == fin -1:
                    i.append(paso1[1])
                    ruta.append(i[-1])
                    costos.append(paso1[0])
                    
                    
def rutasFM(inicio, fin, data):
    inicio = 0
    fin = len(data)
    f = 0
    
    while maxRow(data[inicio])[0] != 0:
        #condiciones iniciales después de la iteración
        i = [0]
        ruta = []
        costos = []

        ruta.append(i[-1])
        paso1 = maxRow(data[i[-1]])
        
        flujoMaximo(ruta, costos, fin, paso1, i)
        
        f += minRow(costos)[0]
        printResults(routeCleaner(ruta), minRow(costos)[0], data)
    
    return f
```

```python
data =[ ['x', 20, 30, 10, 'x'], [0, 'x', 40, 'x', 30], [0,0, 'x', 10, 20], [0, 'x', 5, 'x', 20], ['x', 0, 0, 0, 'x'] ]
#data = [ ['x', 8, 14, 'x', 4], [0, 'x', 5, 7, 6], [0, 10, 'x', 9, 10], ['x', 6, 7, 'x', 5], [0, 0, 0, 0, 'x'] ]

flujo_maximo = rutasFM(0, len(data), data)

print('\nFlujo máximo: ',flujo_maximo)
```
