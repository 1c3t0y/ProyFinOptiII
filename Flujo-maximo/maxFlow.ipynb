```python
def maxRow(arr):
    max_num = 0
    for x in arr:
        if( type(x) == str ):
            continue
        if( x >= max_num ):
            max_num = x
            
    return[max_num, arr.index(max_num)]


def minRow(arr):
    min_num = maxRow(arr)[0]
    for x in arr:
        if( type(x) == str or x == 0):
            continue
        if( x < min_num ):
            min_num = x
            
    return[min_num, arr.index(min_num)]


def printMatrix(arr):
    n = len(arr)
    for i in range(0,n):
        print(arr[i])


def isInArray(arr, x):
    resp = False
    for z in arr:
        if( type(z) == str ):
            continue
        if( z == x ):
            resp = True
        
    return(resp)


def maxRowCompare2(arr1, arr2):
    maxc = maxRow(arr1)
    aux = arr1.copy()
    
    while isInArray(arr2, maxc[1]) == True:
        idc = aux.index(maxc[0])
        aux[idc] = 'x'
        gde = maxRow(aux)[0]
        maxc = [gde, aux.index(gde)]

    return maxc


def inputMatrix():
    from csv import reader
    
    data = []
    chc = input('Ingresar la matriz inicial: \n\n 1) Desde un csv \n 2) Desde terminal \n')
    
    if chc == '1':
        nombre = input('nombre del archivo csv: ')
        # read csv file as a list of lists
        with open(nombre + '.csv', 'r') as read_obj:
            # pass the file object to reader() to get the reader object
            csv_reader = reader(read_obj)
            # Pass reader object to list() to get a list of lists
            data = list(csv_reader)
        
        for i in range(0, len(data[0])):
            for j in range(0, len(data[0])):
                if data[i][j] != 'x':
                    data[i][j] = int(data[i][j])
            
    elif chc == '2':
        nod = int(input('\n Número de nodos en la red: '))
            
        for i in range(0, nod):
            data.append([])
            for j in range(0, nod):
                if i == j:
                    data[i].append('x')
                else:
                    print('Costo para ir del nodo ', i+1, ' al nodo ', j+1,'. Si no hay conexión esciba: x .')
                    cst = input()
                    if cst == 'x':
                        data[i].append('x')
                    else:
                        data[i].append(int(cst))
            
        
    return data


def step2(arr1, arr2):
    result = []
    
    for elem in range(0, len(arr1)):
        
        if type(arr1[elem]) != str:
            if(arr1[elem] > 0 and isInArray(arr2, elem)== False ):
                result.append( (arr1[elem], elem) )
    
    return result


def step3(arr1):
    result = arr1[0]
    
    for x in arr1:
        if x[0] > result[0]:
            result = x
            
    return result


def step4():
    global i
    global data
    
    r = i[-2]
    result = []
    
    for k in range(0, len(data[r])):
        if type(data[r][k]) != str:
            if (data[r][k] > 0 and isInArray(i, k) == False ):
                result.append( (data[r][k], k) ) 
    
    i.pop()
    
    return result
     

def step5(k,i):
    f = []
    for x in k:
        f.append(x[0])
    flujoMax = minRow(f)[0]
    
    global data
    global resultados
    
    for y in range(0, len(i)-1): #proceso para reajustar valores de matriz
        if type(data[i[y]][i[y+1]]) != str:
            data[i[y]][i[y+1]] -= flujoMax
        if type(data[i[y+1]][i[y]]) != str:
            data[i[y+1]][i[y]] += flujoMax
        
    resultados.append((i,flujoMax))
    print('\nLa ruta encontrada es: ', i, ' con flujo máximo de: ', flujoMax)
    printMatrix(data)
    

def step6():
    global resultados
    global FM
    
    for obj in resultados:
        FM += obj[1]
    
    print('\nFlujo Máximo de la red, F =', FM)
    print('\n\n Lista de Rutas: ', resultados)
    
    
def solverFlujoMaximo():
    
    global i
    global n
    global k
    global data
    global errorcatcher
    global resultados
    global FM

    while isInArray(i, n) == False:

        if(errorcatcher > n):
            step6()
            break

        if maxRow(data[0])[0] == 0:
            step6()
            break

        S_i = step2(data[i[-1]], i)

        if len(S_i) != 0 :
            k.append(step3(S_i))
            i.append(k[-1][1])
        else:
            #Posibilidad de fin de ruta
            if(i[-1] == 1):
                step6()
                continue

            #Se calcula de nuevo el S_i tomando un paso atrás
            S_i = step4()
            k.pop()

            if len(S_i) != 0 :
                k.append(step3(S_i))
                i.append(k[-1][1])

            elif len(S_i) == 0 and len(i) == 1:
                print('Dead End')
            else:
                #print('Otra iteración de step4 ?')
                errorcatcher += 1
                S_i = step4()
                k.pop()

                if len(S_i) != 0 :
                    k.append(step3(S_i))
                    i.append(k[-1][1])

            #print('Excepcion:', k, i, S_i)


        #print('Check: ', i, S_i, k)

        if k[-1][1] == n:
            step5(k,i)
            i = [0]
            k = []

            
```
