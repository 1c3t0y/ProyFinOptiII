```python
def maxRow(arr):
    max_num = 0
    for x in arr:
        if( type(x) == str ):
            continue
        if( x >= max_num ):
            max_num = x
            
    return[max_num, arr.index(max_num)]


def minRow(arr):
    min_num = maxRow(arr)[0]
    for x in arr:
        if( type(x) == str or x == 0):
            continue
        if( x < min_num ):
            min_num = x
            
    return[min_num, arr.index(min_num)]


def printMatrix(arr):
    n = len(arr)
    for i in range(0,n):
        print(arr[i])


def isInArray(arr, x):
    resp = False
    for z in arr:
        if( type(z) == str ):
            continue
        if( z == x ):
            resp = True
        
    return(resp)


def maxRowCompare2(arr1, arr2):
    maxc = maxRow(arr1)
    aux = arr1.copy()
    
    while isInArray(arr2, maxc[1]) == True:
        idc = aux.index(maxc[0])
        aux[idc] = 'x'
        gde = maxRow(aux)[0]
        maxc = [gde, aux.index(gde)]

    return maxc


def step2(arr1, arr2):
    result = []
    
    for elem in range(0, len(arr1)):
        
        if type(arr1[elem]) != str:
            if(arr1[elem] > 0 and isInArray(arr2, elem)== False ):
                result.append( (arr1[elem], elem) )
    
    return result


def step3(arr1):
    result = arr1[0]
    
    for x in arr1:
        if x[0] > result[0]:
            result = x
            
    return result


def step4():
    global i
    global data
    
    r = i[-2]
    result = []
    
    for k in range(0, len(data[r])):
        if type(data[r][k]) != str:
            if (data[r][k] > 0 and isInArray(i, k) == False ):
                result.append( (data[r][k], k) ) 
    
    i.pop()
    
    return result
     

def step5(k,i):
    f = []
    for x in k:
        f.append(x[0])
    flujoMax = minRow(f)[0]
    
    global data
    global resultados
    
    for y in range(0, len(i)-1): #proceso para reajustar valores de matriz
        if type(data[i[y]][i[y+1]]) != str:
            data[i[y]][i[y+1]] -= flujoMax
        if type(data[i[y+1]][i[y]]) != str:
            data[i[y+1]][i[y]] += flujoMax
        
    resultados.append((i,flujoMax))
    print('La ruta encontrada es: ', i, ' con flujo máximo de: ', flujoMax)
    printMatrix(data)
    

def step6():
    global resultados
    global FM
    
    for obj in resultados:
        FM += obj[1]
    
    print('F =', FM)
    
```

```python
data =[ ['x', 20, 30, 10, 'x'], [0, 'x', 40, 'x', 30], [0,0, 'x', 10, 20], [0, 'x', 5, 'x', 20], ['x', 0, 0, 0, 'x'] ]
#data = [ ['x', 8, 14, 'x', 4], [0, 'x', 5, 7, 6], [0, 10, 'x', 9, 10], ['x', 6, 7, 'x', 5], [0, 0, 0, 0, 'x'] ]
#data = [['x', 20, 20, 'x', 20, 'x'], [0, 'x', 'x', 'x', 10, 25], [0, 'x', 'x', 20, 15, 'x'], ['x', 'x', 0, 'x', 10, 15], [0, 5, 0, 5, 'x', 40], ['x', 0, 'x', 0, 0, 'x']]

resultados = []
FM = 0

errorcatcher = 0

printMatrix(data)

#Datos iniciales
i = [0]
n = len(data[0]) -1
k = []
```

```python
while isInArray(i, n) == False:
    
    if(errorcatcher > n):
        step6()
        break
    
    if maxRow(data[0])[0] == 0:
        step6()
        break

    S_i = step2(data[i[-1]], i)

    if len(S_i) != 0 :
        k.append(step3(S_i))
        i.append(k[-1][1])
    else:
        #Posibilidad de fin de ruta
        if(i[-1] == 1):
            step6()
            continue

        #Se calcula de nuevo el S_i tomando un paso atrás
        S_i = step4()
        k.pop()

        if len(S_i) != 0 :
            k.append(step3(S_i))
            i.append(k[-1][1])

        elif len(S_i) == 0 and len(i) == 1:
            print('Dead End')
        else:
            #print('Otra iteración de step4 ?')
            errorcatcher += 1
            S_i = step4()
            k.pop()
            
            if len(S_i) != 0 :
                k.append(step3(S_i))
                i.append(k[-1][1])

        #print('Excepcion:', k, i, S_i)


    #print('Check: ', i, S_i, k)

    if k[-1][1] == n:
        step5(k,i)
        i = [0]
        k = []
```
