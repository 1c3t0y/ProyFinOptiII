```python
data =[ ['x', 20, 30, 10, 'x'], [0, 'x', 40, 'x', 30], [0,0, 'x', 10, 20], [0, 'x', 5, 'x', 20], ['x', 0, 0, 0, 'x'] ]

def maxRow(arr):
    max_num = 0
    for x in arr:
        if( type(x) == str ):
            continue
        if( x > max_num ):
            max_num = x
            
    return[max_num, arr.index(max_num)]

def minRow(arr):
    min_num = maxRow(arr)[0]
    for x in arr:
        if( type(x) == str or x == 0):
            continue
        if( x < min_num ):
            min_num = x
            
    return[min_num, arr.index(min_num)]

def printMatrix(arr):
    n = len(arr)
    for i in range(0,n):
        print(arr[i])


def isInArray(arr, x):
    resp = False
    for z in arr:
        if( type(z) == str ):
            continue
        if( z == x ):
            resp = True
        
    return(resp)

def maxRowCompare(arr1, arr2):
    maxc = maxRow(arr1)
    
    while isInArray(arr2, maxc[1]) == True:
        aux = arr1.copy()
        aux.remove(maxc[0])
        maxc = maxRow(aux)
    return(maxc[0], arr1.index(maxc[0]) )
        

printMatrix(data)

inicio = 0
fin = len(data)

#condiciones iniciales del proceso
ruta = []
costos = []
i = 0
ruta.append(i)
paso1 = maxRow(data[i])
costos.append(paso1[0])

#Se itera siempre que haya un valor en la fila inicial mayor a cero

while maxRow(data[inicio])[0] > 0:

    while isInArray(ruta, fin-1) == False:
        if not isInArray(ruta, paso1[1]) :
            #print('no esta en el arreglo, podemos continuar',)
            i = paso1[1]
            ruta.append(i)
            costos.append(paso1[0])
            paso1 = maxRowCompare(data[i], ruta)
            print('---------paso1:',paso1)
            if paso1[0] == 0:
                continue;
        else:
            print('si esta en el arreglo, buscamos otro paso')


    costo_min = minRow(costos)[0]
    print('Iteración -------------------------------------------------------------------')
    print('La ruta más corta es: ',ruta ,' el flujo máximo de esta ruta es:', costo_min)

    for y in range(0, len(ruta)-1): #proceso para reajustar valores de matriz
        data[ruta[y]][ruta[y+1]] = data[ruta[y]][ruta[y+1]] - costo_min
        data[ruta[y+1]][ruta[y]] = data[ruta[y+1]][ruta[y]] + costo_min

    printMatrix(data)

    #condiciones iniciales después de la iteración
    i = 0
    ruta = []
    costos = []
    ruta.append(i)
    paso1 = maxRow(data[i])
    costos.append(paso1[0])
```
